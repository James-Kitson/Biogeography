**MCC Tree selection from MrBayes files**

The following code will take your MrBayes tree files, read a specified number of lines which covers the taxablock (in this case [:63]) and then write the last 75% of your trees from each tree file to the same file (specified by gen >= 25000000 for a 100000000 generation analysis like this one) The resulting files can be run through [TreeAnnotator (part of the BEAST package)](http://beast.bio.ed.ac.uk/downloads) to select MCC trees for character reconstruction, biogeography etc.

**Code with paths for calibrated trees**

```python
hndl = open('../MrBayes/Tree_files/MrBayes_treesets/Croc_test/Croc_Bayes.trees','wt')

lines = open('../MrBayes/Tree_files/MrBayes_treesets/Croc_test/Croc_Bayes_Nov16_unordered_unconstrained_70.nex.run1.t','r').readlines()[:102]
for l in lines:
    hndl.write(l)
    print l
    
from glob import glob

for f in glob('../MrBayes/Tree_files/MrBayes_treesets/Croc_test/Croc_Bayes_Nov16_unordered_unconstrained_70.nex.run*'):
    with open(f,'r') as lines:
        for l in lines:
            if l.startswith('   tree '):
                gen = int(l.split('.')[1].split('[')[0])
                if gen >= 25000000:
                    hndl.write(l)   
                    
hndl.write('end;\n')
hndl.close()   
```
